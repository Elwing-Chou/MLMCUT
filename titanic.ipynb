```
import pandas as pd
train_df = pd.read_csv("/gdrive/My Drive/titanic/train.csv", encoding="utf-8")
train_df
```

```
test_df = pd.read_csv("/gdrive/My Drive/titanic/test.csv", encoding="utf-8")
test_df
```

```
train_df.isna().sum()
```

```
med = train_df.median()
train_df = train_df.fillna(med)
train_df.isna().sum()
```

```
test_df.isna().sum()
```

```
# ä¸è¦å†é‡ç®—ä¸­ä½æ•¸
test_df = test_df.fillna(med)
test_df.isna().sum()
```

```
most_freq = train_df["Embarked"].value_counts().idxmax()
train_df["Embarked"] = train_df["Embarked"].fillna(most_freq)
train_df.isna().sum()
```

```
# ä¸è¦å†é‡æ–°å¾—åˆ°æœ€å¸¸å‡ºç¾
test_df["Embarked"] = test_df["Embarked"].fillna(most_freq)
test_df.isna().sum()
```

```
dummy = pd.get_dummies(train_df["Embarked"])
train_df = pd.concat([train_df, dummy], axis=1)
train_df
```

```
dummy = pd.get_dummies(test_df["Embarked"])
test_df = pd.concat([test_df, dummy], axis=1)
```

```
dummy = pd.get_dummies(train_df["Sex"])
train_df = pd.concat([train_df, dummy], axis=1)
```

```
dummy = pd.get_dummies(test_df["Sex"])
test_df = pd.concat([test_df, dummy], axis=1)
```

```
# Series(è¡Œ).apply(æµç¨‹)
def middle(s):
    return s.split(",")[-1].split(".")[0].replace(" ", "")
mid = train_df["Name"].apply(middle)
pd.crosstab(mid, train_df["Survived"]).T
```

```
# Series(è¡Œ).apply(æµç¨‹)
def middle(s):
    reserved = ["Master", "Miss", "Mr", "Mrs"]
    mid = s.split(",")[-1].split(".")[0].replace(" ", "")
    if mid in reserved:
        return mid
    else:
        return "X"
mids = train_df["Name"].apply(middle)
dummy = pd.get_dummies(mids)
train_df = pd.concat([train_df, dummy], axis=1)
train_df
```

```
mids = test_df["Name"].apply(middle)
dummy = pd.get_dummies(mids)
test_df = pd.concat([test_df, dummy], axis=1)
test_df
```

```
train_df = train_df.drop(['PassengerId', 'Name', 'Sex',
               'Ticket', 'Cabin', 'Embarked', 'X'], axis=1)
train_df
```

```
test_df = test_df.drop(['Name', 'Sex',
               'Ticket', 'Cabin', 'Embarked', 'X'], axis=1)
test_df
```
